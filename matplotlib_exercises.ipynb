```python
# import libraries / modules
# man I hope this works

import matplotlib.pyplot as plt

# other libraries

import numpy as np
import math as m
from random import randint
import random
import pandas as pd
```

```python
x = range(-50, 51)
y = [n**2 - n + 5 for n in x]

plt.annotate('origin', xy=(0,0), xytext=(0,300), arrowprops={'facecolor': 'blue'})

plt.plot(x,y)
plt.show()
```

```python
# 2a. create a plot that represents y = sqrt of x

plt.figure(figsize = (10,6))
plt.grid(True, ls = ':')

x = range(0,51)
y = [m.sqrt(n) for n in x]
plt.title('$y=\sqrt{x}$', fontsize = 14)
plt.xlabel('$x$', fontsize = 12)
plt.ylabel('$\sqrt{x}$', fontsize = 12, rotation = 0)
plt.plot(x,y)
plt.show()
```

```python
# 2b. create a plot that represents y = x^3

plt.figure(figsize = (10,6))
plt.grid(True, ls = ':')

x = range(-50,51)
y = [n**3 for n in x]
plt.title('$y = x^3$', fontsize = 14)
plt.xlabel('$x$', fontsize = 12)
plt.ylabel('$x^3$', fontsize = 12, rotation = 0)
plt.plot(x,y)
plt.show()
```

```python
# 2c. create and label a plot that represents y = 2^x

plt.figure(figsize = (10,6))

x = range(-50,51)
y = [2**n for n in x]

plt.plot(x,y)
plt.grid(True, ls = ':')
plt.title('$y = 2^x$', fontsize = 14)
plt.xlabel('$x$', fontsize = 12)
plt.ylabel('$2^x$', fontsize = 12, rotation = 0)
plt.plot(x,y)
plt.show()
```

```python
# 2d. create and label a plot that represents y = 1/(x+1)

plt.figure(figsize = (10,6))
plt.grid(True, ls = ':')

x = range(0,51)
y = [1/(n+1) for n in x]
plt.subplot(2,2,1)
plt.title('$y=1/(x+1)$', fontsize = 14)
plt.xlabel('$x$', fontsize = 12)
plt.ylabel('$1/(x+1)$', fontsize = 12, rotation = 0)
plt.plot(x,y)
plt.show()
```

```python
# 3. Move those four charts into a subplot

x = [x * 0.25 for x in range(0, 51)]

y1 = [m.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [(1/(n+1)) for n in x]

plt.figure(figsize = (10,8))

plt.suptitle('One large Figure w/ 4 Subplots')

plt.subplot(2,2,1)
plt.title('$y=\sqrt{x}$', fontsize = 14)
plt.xlabel('$x$', fontsize = 12)
plt.ylabel('$\sqrt{x}$', fontsize = 12)
plt.grid(True, ls = ':')
plt.plot(x,y1, ls = '-', color = 'red')

plt.subplot(2,2,2)
plt.title('$y=x^3$', fontsize = 14)
plt.xlabel('$x$', fontsize = 12)
plt.ylabel('$x^3$', fontsize = 12)
plt.grid(True, ls = ':')
plt.plot(x,y2, ls = '-', color = 'blue')

plt.subplot(2,2,3)
plt.title('$y=2^x$', fontsize = 14)
plt.xlabel('$x$', fontsize = 12)
plt.ylabel('$2^x$', fontsize = 12)
plt.grid(True, ls = ':')
plt.plot(x,y3, ls = '-', color = 'green')

plt.subplot(2,2,4)
plt.title('$y=1/(x+1)$', fontsize = 14)
plt.xlabel('$x$', fontsize = 12)
plt.ylabel('$1/(x+1)$', fontsize = 12)
plt.grid(True, ls = ':')
plt.plot(x,y4, ls = '-', color = 'yellow')

plt.tight_layout()
plt.show()


```

```python
# Combine the figures you created in the last step into one figure where each of the 4 
# equations has a different color for the points. Be sure to include a legend and an 
# appropriate title for the figure.




# series
x = [x*.25 for x in range(0,10)]

y1 = [m.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n+1) for n in x]

# sqrt plot
plt.plot(x, y1, color = 'red', label = 'sqrt')

# cube plot
plt.plot(x, y2, color = 'blue', label = 'cube')

# 2 to the x power
plt.plot(x, y3, color = 'green', label = '2^x')

# 1/(x+1)
plt.plot(x, y4, color = 'yellow', label = '1/(x+1)')

# Chart labels/sizing/etc
plt.title('Four plots in One')
plt.xlabel('x')
plt.ylabel('Formulas')
plt.grid(True, ls = '--')
plt.legend()
plt.show()


```
