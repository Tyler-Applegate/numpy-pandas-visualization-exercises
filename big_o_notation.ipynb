```python
# import necessary libraries

import matplotlib.pyplot as plt
import numpy as np
import math as m
```

```python
# Entering all necessary equations

x = range(1,4)

y1 = [0*n + 1 for n in x]
y2 = [m.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * m.log(n) for n in x]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [m.factorial(n) for n in x]
y8 = [n**n for n in x]
```

```python
# Make a new Jupyter notebook named big_o_notation.ipynb
# import necessary libraries

# Title your chart "Big O Notation"
# Label your x axis "Elements"
# Label your y axis "Operations"
# Label your curves or make a legend for the curves
# Use LaTex notation where possible

# setting up my labels and dimensions
plt.figure(figsize = (10,8))
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')



# entering all of my equations

plt.plot(x, y1, ls = '-', label = '$O(1)$')
plt.plot(x, y2, ls = '-', label = '$O(log n)$')
plt.plot(x, y3, ls = '-', label = '$O(n)$')
plt.plot(x, y4, ls = '-', label = '$O(n log n)$')
plt.plot(x, y5, ls = '-', label = '$O(n^2)$')
plt.plot(x, y6, ls = '-', label = '$O(2^n)$')
plt.plot(x, y7, ls = '-', label = '$O(n!)$')
plt.plot(x, y8, ls = '-', label = '$O(n^n)$')
plt.legend()
plt.show()
```

```python
# Here is how we did Big O during review...
# would help if we kne
x = range(1,20)

y1 = [1 for n in x]
y2 = [m.log(n) for n in x]
y3 = x
y4 = [(n * m.log(n)) for n in x]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [m.factorial(n) for n in x]
y8 = [n**n for n in x]

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operation')

plt.plot(x,y1, color='cyan', label='SO(1)$')
plt.plot(x,y2, color='red', label='O(log n)$')
plt.plot(x,y3, color='b', label='$O(n)$')
plt.plot(x,y4, color='g', label='$O(n log n)$')
plt.plot(x,y5, color='coral', label='$O(n^2)$')
plt.plot(x,y6, color='olive', label='$O(n^2)$')
plt.plot(x,y7, color='lime', label='$O(n!)$')
plt.plot(x,y8, color='violet', label='$O(n^n)$')
         
# plt.xlim(1,11)
# plt.ylim(0,50)

plt.legend(loc='upper right', fontsize=14)
plt.show()
         
```
